A reologia é um campo da física preocupado com o fluxo e alterações da forma da matéria, sejam sólidos, líquidos ou gases em função de um comportamento dependente do tempo sob a influência de tensões. O principal foco está em substancias que não possuem uma viscosidade ou estado de fluxo continuo.

O reômetro é um aparelho utilizado para medições reológicas, no qual pode ser definido como um tipo de viscosímetro capaz de medir as propriedades visco-elásticas dos materiais além da sua viscosidade, mede também o comportamento do material, como tensão de escoamento, propriedades cinéticas, viscosidade complexa, módulo, fluência e recuperação.

Essas medidas podem ajudar a prever o prazo de validade dos produtos sob várias condições de estresse. Além disso, pode ser utilizado em demais áreas, assim como, na área médica.

## Introdução

A reologia vem dO termo derivado da palavra grega "rheo", que significa "fluir", então significa literalmente "a medição do fluxo", no entanto, experimentos reológicos revelam não apenas as propriedades de fluxo de líquidos, mas também as propriedades de deformação de sólidos. A conexão entre eles é que o material fluirá enquanto houver grandes forças de deformação tangencial. Essa deformação tangencial é chamada de cisalhamento.

A reologia pode ser usada para descrever vários tipos de comportamento de cisalhamento, e os materiais geralmente se situam entre dois extremos: fluidos viscosos ideais (como óleo mineral) e sólidos elásticos ideais (como vigas de aço). Quando esses comportamentos ideais são combinados, diz-se que o material exibe viscoelasticidade e pode estar mais próximo de qualquer um dos lados do espectro.

Reometria é o termo usado para descrever técnicas usadas para determinar dados reológicos.Tanto os sólidos quanto os líquidos podem ser estudados usando reômetros oscilatórios e rotacionais. Testes rotacionais são usados ​​para caracterizar a viscosidade, testes de oscilação e testes de relaxamento e fluência são usados ​​para caracterizar a viscoelasticidade.

O modelo de placas paralelas é um dispositivo ilustrativo para definir parâmetros reológicos básicos. É composto por duas placas, uma abaixo e outra acima, entre as quais a área (A) e a altura (h) são constantes e o material preenche o espaço entre as placas. Quando uma força tangencial (F) é aplicada, o material se deforma.

##Cassificação Reológica


Em termos de deformação, os fluidos podem ser divididos em:
- Reversíveis ou elásticos: são sistemas não drenados; sua deformação é reversível e o sistema obedece à lei de Hooke.
- irreversíveis ou viscosos: são sistemas fluidos; sua deformação é
Irreversível, o sistema segue a lei de Newton da viscosidade constante.
 A classificação também pode ser baseada na relação entre proporções
Deformação e tensão de cisalhamento:
- Fluidos newtonianos: sua viscosidade é constante e seguem as leis de Newton. Esta categoria abrange todos os gases e líquidos não poliméricos e homogêneos.
Exemplos são água, leite, solução de sacarose, óleo vegetal.
- Fluidos não newtonianos: a relação entre taxa de deformação e tensão de cisalhamento não é constante.
 Além disso, os fluidos não newtonianos ainda podem ser classificados
em: viscoelástico, dependente do tempo e independente do tempo.

A tixotropia é explicada pela diminuição da viscosidade com o tempo que uma taxa de cisalhamento é aplicada à amostra. Já para a reologia ocorre o contrário, com o tempo de aplicação da taxa de cisalhamento sua viscosidade aumenta. Esse comportamento é dependente do tempo, ou seja, após um certo período de observação, a amostra volta à sua viscosidade inicial, caso isso não ocorra, observa-se uma fluidização ou espessamento irreversível da amostra.

##Comportamento de materiais viscoelásticos

A viscoelasticidade consiste em componentes elásticos e viscosos que a maioria dos materiais possui. Qual componente é dominante pode ser determinado por testes reológicos oscilatórios dinâmicos. Essa caracterização foi feita com testes de oscilação dinâmica, onde as diferenças entre materiais com comportamento mais elástico do que viscoso se devem ao tempo que a amostra leva para se deformar entre as oscilações da tensão aplicada ao dispositivo, resultando em um ângulo de fase maior que zero grau . Assim, atrasos de resposta de amostra mais longos correspondem a componentes viscosos maiores, enquanto tempos de atraso curtos correspondem a materiais com componentes elásticos mais altos. Sabendo disso, a caracterização reológica pode ser feita avaliando mudanças na tensão de cisalhamento e na frequência de oscilação, além de modificar outras condições como temperatura e tempo de análise.

##Varredura de Frequência/Tempo

G'<G".
Esse teste permite observar o comportamento da microestrutura do material com o

##Curva de escoamento


É avaliado o comportamento da tensão e viscosidade em função da taxa de deformação, em regime permanente. Os fluidos analisados seguem o comportamento Herschel-Bulkley, para fluido com tensão limite de escoamento, e viscosidade caindo com a taxa de deformação.
Nesse teste, para que seja feito de modo relativamente mais rápido, avalia-se a curva a partir de taxas de deformação mais altas para mais baixas, pois desta forma, o tempo para atingir o regime permanente é menor, visto que a amostra sofre um pré-cisalhamento. Começando de taxas mais baixas para mais altas, o teste demoraria muito mais tempo para ser executado.

##Tensão constante


O fluido é submetido a uma tensão constante durante um tempo definido e assim é
avaliado o comportamento da taxa de cisalhamento. Dessa forma, é possível obter com precisão o valor da tensão limite de escoamento e avaliar efeitos tixotrópicos próximos a essa  região.
Nesse teste, o valor encontrado para a tensão limite é mais preciso em comparação com o valor obtido pela curva de escoamento. Isso ocorre, porque o teste da curva de escoamento inicia-se das maiores taxas para as menores, ou seja, a estrutura é pré-cisalhada e propicia medidas menores da tensão crítica, enquanto que no creep (tensão constante) isso não ocorre, tornando o resultado mais preciso.

##Degrau de tensão

Esse teste parte do mesmo princípio que o Creep, porém ele é realizado em dois passos. Primeiramente, submete-se o fluido a uma tensão x durante certo tempo, e em seguida o submete a uma tensão y durante outro período de tempo. Após isso, o caminho inverso também é realizado. Dessa forma é possível visualizar o comportamento do fluido précisalhado a diferentes tensões, ou seja, a construção e reconstrução da microestrutura. É possível verificar nesse teste se o fluido apresenta comportamento viscoelástico e efeitos tixotrópicos.
